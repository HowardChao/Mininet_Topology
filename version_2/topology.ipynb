```python
# def parse_file():
line_list = []
with open('connection_100_v2.csv') as f:
    for each_line in f: 
        if each_line[0] == '#':
            print(each_line)
            continue
        elif each_line == 'c\n':
            break
        else:
            a = each_line.split(",")
            line_list.append(a)
            print(a)
bottom_start_id = int(line_list[0][0])
switch_num = int(line_list[0][1])
bandwidth = int(line_list[0][2])
```

```python
class LinuxRouter( Node ):
    "A Node with IP forwarding enabled."

    def config( self, **params ):
        super( LinuxRouter, self).config( **params )
        # Enable forwarding on the router
        self.cmd( 'sysctl net.ipv4.ip_forward=1' )

    def terminate( self ):
        self.cmd( 'sysctl net.ipv4.ip_forward=0' )
        super( LinuxRouter, self ).terminate()
```

```python
def runNet():
    "Create and run the network"
    info( "*** Starting network\n" )
    topo = custom_topo()
    net = Mininet( topo = topo, controller=Controller, switch=OVSSwitch )
#     c1 = Controller( 'c0', port=6633 )
#     net.addController( controller = c1)
    net.start()
    info( '*** Starting controllers\n')
    for controller in net.controllers:
        controller.start()

    info( '*** Starting switches\n')
    for x in range(106):
        net.get('s'+str(x)).start('c0')
    CLI(net)
```
